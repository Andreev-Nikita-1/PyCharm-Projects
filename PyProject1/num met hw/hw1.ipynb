```python
import numpy as np
import matplotlib.pyplot as plt
```

$\frac{1}{k^2-k-z} \leq \frac{1}{k^2-k-2}$

$\sum\limits_{n+1}^\infty \frac{1}{k^2-k-z} \leq  
\sum\limits_{n+1}^\infty \frac{1}{k^2-k-2} \leq 
\int\limits_n^\infty\frac{1}{x^2-x-2} =
-\frac{1}{3}( \log(n-2) - \log(n+1)) <
10^{-6}$

$\log\frac{n+1}{n-2} < 3\cdot10^{-6}$

$e^{3\cdot 10^{-6}}-1 \approx 3\cdot 10^{-6}$

$\frac{n+1}{n-2} <
3\cdot 10^{-6} +1<
e^{3\cdot 10^{-6}}$

$a = 3\cdot 10^{-6}+1$

$n+1 < an - 2a$

$n>\frac{2a+1}{a-1}$

$n>10^6+2$

```python
def w(z):
    return np.sum([1/(k**2-k-z) for k in range(1000002, 0, -1)])
```

```python
def plot(xs, ys):
    plt.plot(xs, ys)
    plt.show()
    
```

```python
xs = np.arange(0.1, 1.9+2/100, 2/100)
ys = [w(x) for x in xs]
```

```python
plot(xs, ys)
```

$\sum\limits_2^\infty\frac{1}{k^2-k} = 1$

$k^2-k-2>\sqrt{0.95}k^2, k>41$

$\frac{1}{k^2-k-z} - \frac{1}{k^2-k} =
\frac{z}{(k^2-k-z)(k^2-k)} < 
\frac{z}{0.95 k^4}$

$\sum\limits_{n+1}^\infty \frac{1}{k^4} < 
\int\limits_n^\infty \frac{1}{x^4} = 
\frac{1}{3n^3}$

$\frac{z}{0.95\cdot3n^3} < 10^{-6}$

$n>88$

```python
def wb(z):
    return 1 - 1/z + z*np.sum([1/((k**2-k)*(k**2-k-z)) for k in range(88, 1, -1)])
```

```python
xs = np.arange(0.1, 1.9+2/100, 2/100)
ys = [wb(x) for x in xs]
```

```python
plot(xs, ys)
```

```python
xs = np.arange(0.1, 1.9+2/100, 2/100)
ys = [wb(x)-w(x) for x in xs]
```

```python
plot(xs, ys)
```
