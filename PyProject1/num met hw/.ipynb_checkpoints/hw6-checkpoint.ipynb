```python
import numpy as np
import matplotlib.pyplot as plt
```

$A_{i,j} = \int\limits_0^1x^i x^j
=\frac{1}{i+j+1}$

```python
def get_matrix(n):
    return np.array([[1 / (i + j + 1) for i in range(n)] for j in range(n)])
```
