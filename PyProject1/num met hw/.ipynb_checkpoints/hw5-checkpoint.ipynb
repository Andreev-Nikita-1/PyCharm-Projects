```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
import scipy.special
```

# A

$w(x) = 
\sum\limits_3^\infty\frac{1}{k^2-k-x}$

$I = \int\limits_0^3 w(x) d x$ 

$\int\frac{1}{k^2-k-x} =
\log \left(\frac{k^2 - k}{k^2-k-3}\right) =
\log \left(1 +\frac{3}{k^2-k-3}\right) = 
\frac{3}{k^2-k-3} +O(\frac{1}{k^4}) \approx 
\frac{3}{k^2}$

$\log \left(1 +\frac{3}{k^2-k-3}\right) - 
\frac{3}{k^2} - \frac{3}{k^3}=
O(\frac{1}{k^4})$

$\sum\limits_{N+1}^\infty 
\log \left(1 +\frac{3}{k^2-k-3}\right) - 
\frac{3}{k^2} - \frac{3}{k^3}=
O(\frac{1}{N^3})$


```python
def term_int(k):
    return (np.log((k**2 - k) / (k**2 - k - 3)) - 3 / k**2) 
```

```python
def series_int():
    # N=1e+6, error ~= 1e-18
    return np.sum([term_int(k) for k in range(3, int(1e+6))][::-1]) - np.sum([
        3 / k**3 for k in range(3, int(1e+6))
    ][::-1]) + np.pi**2 / 2 + 3 * scipy.special.zeta(3) - 6 - 3 / 4 - 3 / 8
```

```python
true_int = series_int()
print(r'"reference" value = {}'.format(true_int))
```

# B

$\sum\limits_3^\infty
\frac{1}{k^2-k-x} = 
\frac{\pi^2}{6} - 1 - \frac{1}{4} +
\zeta(3) - 1 - \frac{1}{8} +
(\zeta(4) - 1 - \frac{1}{16})(1+x) +
\sum\limits_3^\infty
\left(\frac{1}{k^2-k-x} -
\frac{1}{k^2} -
\frac{1}{k^3} - 
\frac{1+x}{k^4}\right)$

$\sum\limits_3^\infty
\left(\frac{1}{k^2-k-x} -
\frac{1}{k^2} -
\frac{1}{k^3} - 
\frac{1+x}{k^4}\right) =
\frac{k(2x+1) + (1+x)x}{k^4(k^2-k-x)} = O(k^5)$

$\sum\limits_{N+1}^\infty
\frac{k(2x+1) + (1+x)x}{k^4(k^2-k-x)}=
O(\frac{1}{N^4})$

```python
def w(x, N=int(1e+4)):
    # N = 1e+4, error ~= 1e-16
    def term(k):
        return (k * (1 + 2 * x) + (1 + x) * x) / (k**4 * (k**2 - k - x))

    const = np.pi**2 / 6 + scipy.special.zeta(
        3) - 2 - 1 / 4 - 1 / 8 + (1 + x) * (scipy.special.zeta(4) - 1 - 1 / 16)
    return const + np.sum([term(k) for k in range(3, N + 1)][::-1])
```

```python
def trapeze(f, N):
    h = 3 / N
    return h * ((f(0) + f(3)) / 2 + np.sum([f(n * h) for n in range(1, N)]))
```

```python
def simpson(f, N):
    h = 3 / N

    def coeff(k):
        if k == 0 or k == N:
            return 1
        elif k == int(k):
            return 2
        else:
            return 4

    return h / 6 * np.sum(
        [coeff(k) * f(k * h) for k in np.arange(0, N + 1 / 2, 1 / 2)][::-1])
```

```python
Ns = np.array([int(np.power(10, nlog)) for nlog in np.arange(0, 3, 0.1)])
trapezes = [np.log10(np.abs(trapeze(w, n) - true_int)) for n in Ns]
simpsons = [np.log10(np.abs(simpson(w, n) - true_int)) for n in Ns]
pylab.rcParams['figure.figsize'] = 15, 7
pylab.plot(np.log10(Ns + 1), trapezes, label="trapeze")
pylab.plot(np.log10(Ns + 1), simpsons, label="simpson")
pylab.legend()
pylab.grid()
```

При вычислении функции выбираю точность примерно равную машинному эпсилон (1e-16). На графике видно, что, действительно, наклон прямой для метода трапеций равен примерно 2, а для метода Симпсона - 4, что согласуется с теоретическими оценками.

# C

```python
def newton(deg, a, b, eps=1e-15):
    f = scipy.special.legendre(deg)
    #     df = f.deriv(1)
    #     почему то начинает считать неправильно при степенях порядка 37
    df = lambda z: deg / (1 - z**2) * (scipy.special.legendre(deg - 1)
                                       (z) - z * f(z))
    x = (a + b) / 2
    while True:
        fx = f(x)
        x1 = x - fx / df(x)
        if x1 < a or x1 > b:
            fa, fb = f(a), f(b)
            if fx * fa < 0:
                x1 = (a + x) / 2
            elif fx * fb < 0:
                x1 = (b + x) / 2
        if np.abs(x - x1) < eps:
            break
        x = x1
    return x
```

```python
def find_roots(deg, method=newton):
    roots_legendre = {0: np.array([])}
    for d in range(1, deg + 1):
        roots_bounds = np.hstack([[-1], roots_legendre[d - 1], [1]])
        roots = []
        for a, b in zip(roots_bounds, roots_bounds[1:]):
            roots.append(method(d, a, b))
        roots_legendre[d] = np.array(roots)
    return roots_legendre
```

```python
def weights(deg, xs):
    f = scipy.special.legendre(deg)
    # если считать производную численно или через f.deriv(1), 
    # то при достижении точности  порядка 1e-10 что-то ломается и результат не улучшается
    df = lambda z: deg / (1 - z**2) * (scipy.special.legendre(deg - 1)
                                       (z) - z * f(z))
    ans = [2 / (1 - x**2) / df(x)**2 for x in xs]
    return ans
```

```python
def gauss(f, zs, ws):
    fs = np.array([f(z) for z in zs])
    return fs.dot(ws)
```

```python
Ns = np.array([int(np.power(10, nlog)) for nlog in np.arange(0, 2, 0.05)])
trapezes = [np.log10(np.abs(trapeze(w, n) - true_int)) for n in Ns]
simpsons = [np.log10(np.abs(simpson(w, n) - true_int)) for n in Ns]
knots = find_roots(Ns[-1])
gausses = [
    np.log10(
        np.abs(3 / 2 * gauss(w, 3 / 2 * (1 + knots[n]), weights(n, knots[n])) -
               true_int)) for n in Ns
]
pylab.rcParams['figure.figsize'] = 15, 7
pylab.plot(np.log10(Ns + 1), trapezes, label="trapeze")
pylab.plot(np.log10(Ns + 1), simpsons, label="simpson")
pylab.plot(np.log10(Ns), gausses, label="gauss")
pylab.xlabel("log10(n)")
pylab.ylabel("log10(error)")
pylab.legend()
pylab.grid()
```

Формула из интернета для ошибки:

$R \asymp ||f^{(2n)}||(\frac{b-a}{3n})^{2n}\frac{1}{\sqrt n} \approx
||f^{(2n)}|| n^{-2n-\frac{1}{2}} C^n$

$||f^{(2n)}|| = \sum\limits_3^\infty \frac{(2n)!}{(k^2-k-3)^{(2n+1)}} \approx n^{2n+\frac{1}{2}} C^n$

$R \asymp C^n$

Проверим, что сходится:

```python
Ns = np.arange(1, 10)
knots = find_roots(Ns[-1])
gausses = [
    np.log10(
        np.abs(3 / 2 * gauss(w, 3 / 2 * (1 + knots[n]), weights(n, knots[n])) -
               true_int)) for n in Ns
]
pylab.rcParams['figure.figsize'] = 15, 7
pylab.plot(Ns, gausses, label="gauss")
pylab.xlabel("n")
pylab.ylabel("log10(error)")
pylab.legend()
pylab.grid()
```
